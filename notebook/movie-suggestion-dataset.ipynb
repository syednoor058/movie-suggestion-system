### Import all the important libraries

```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv('data_sets/tmdb_5000_movies.csv')
credits_df = pd.read_csv('data_sets/tmdb_5000_credits.csv')
```

```python
movies_df.head(5)
```

```python
credits_df.head(5)
```

### Merging the two dataframes into one according to its 'title' tag

```python
movies_df = movies_df.merge(credits_df, on='title')
```

```python
movies_df.head(5)
```

### Analyse the dataframe and clustering process. Sorting the info columns which are actually important for this project

```python
movies_df = movies_df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies_df.head(5)
```

### Removing all the nul values

```python
movies_df.dropna(inplace=True)
```

### Collecting the necessary tags from the 'genre' & 'keywords' columns and converting them into a list of string using a helping function

```python
import ast
def tag_collector(column_tag):
    tag_list = []
    for i in ast.literal_eval(column_tag):
        tag_list.append(i['name']) 
    return tag_list
```

```python
movies_df['genres'] = movies_df['genres'].apply(tag_collector)
movies_df['keywords'] = movies_df['keywords'].apply(tag_collector)
```

### Collecting the top 3 casts from the 'cast' column and converting them into a list of string using a helping function

```python
def cast_collector(column_tag):
    cast_list = []
    counter = 0
    for i in ast.literal_eval(column_tag):
        if counter < 3:
            cast_list.append(i['name'])
        counter+=1
    return cast_list 
```

```python
movies_df['cast'] = movies_df['cast'].apply(cast_collector)
```

```python
def dir_collector(column_tag):
    dir_list = []
    for i in ast.literal_eval(column_tag):
        if i['job'] == 'Director':
            dir_list.append(i['name'])
    return dir_list 
```

```python
movies_df['crew'] = movies_df['crew'].apply(dir_collector)
```

### Remove spaces to make the tag properly work

```python
def space_remover(column_list):
    new_column_list = []
    for i in column_list:
        new_column_list.append(i.replace(" ",""))
    return new_column_list
```

```python
movies_df['cast'] = movies_df['cast'].apply(space_remover)
movies_df['crew'] = movies_df['crew'].apply(space_remover)
movies_df['genres'] = movies_df['genres'].apply(space_remover)
movies_df['keywords'] = movies_df['keywords'].apply(space_remover)
```

### Split the description into tags

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

### Combining all the tags in one column and drop those columns

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
movies_df = movies_df.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
movies_df.head(5)
```

### Joining all the tags together

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x: " ".join(x))
movies_df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
movie_tag_vector = cv.fit_transform(movies_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
movie_similarity = cosine_similarity(movie_tag_vector)
```

```python
def suggestion(movie):
    index = movies_df[movies_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(movie_similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies_df.iloc[i[0]].title)
```

```python
suggestion('The Dark Knight Rises')
```
